```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4]
y = [108, 528, 3.3, 10.9]
#y = np.log(y)
plt.bar(x, y, width=0.4)

label = ['104 min','528 min','3 min','11 min']
# Add text on top of each bar
for i in range(len(x)):
    plt.text(x[i], y[i] + 0.2, label[i], ha='center',fontsize=14)

font_g = 14

plt.xlabel('Methods', fontsize=font_g)
plt.ylabel('Computational time / min', fontsize=font_g)
#plt.title('Computational time compare with FEM and DNNs')
lables = ['FEM \n 5 mm mesh \n 1 million DOFs','FEM \n 3 mm mesh \n 4.7 million DOFs',
          'DNNs\n data shape \n 4×1000 ','DNNs \n data shape \n  4×4000']
plt.xticks(x, lables, fontsize=12)

plt.tight_layout()
#plt.grid(True)
plt.savefig('t_compare.pdf')
```
