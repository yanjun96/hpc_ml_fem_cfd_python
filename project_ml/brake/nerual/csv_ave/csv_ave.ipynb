# CSV to CSV_ave

```python
import os
import pandas as pd

source_dir1 = "/Users/yanjun/Documents/pdc_fem_cfd_python_cpp/project/brake/data/csv/"
target_dir = "/Users/yanjun/Documents/pdc_fem_cfd_python_cpp/project/brake/nerual/csv_ave/"

csv_files = [f for f in os.listdir(source_dir1) if f.endswith(".csv")]

for file in csv_files:
    file_path = os.path.join(source_dir1, file)
    df = pd.read_csv(file_path)
    df.columns = df.iloc[0]
    df = df[1:].reset_index(drop=True)
    df = df.iloc[:, :3]
    df.columns = ["t1", "t2", "t3"]
    df = df.astype({'t1': float, 't2': float, 't3': float})
    df['ave'] = (df['t1'] + df['t2'] + df['t3']) / 3
    new_f_name = os.path.join(target_dir, file)
    df.to_csv(new_f_name, index=False)
```
