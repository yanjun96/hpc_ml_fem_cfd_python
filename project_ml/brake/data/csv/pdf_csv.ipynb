```python
import os
import pdfplumber
import pandas as pd

source_dir = "/Users/yanjun/Documents/pdc_fem_cfd_python_cpp/project/brake/data/csv/"
pdf_files = [f for f in os.listdir(source_dir) if f.endswith(".pdf") ]
print(pdf_files)
data = []
n=0
for file in pdf_files:
   data = []
   with pdfplumber.open(file) as pdf:
     for page in pdf.pages:
         table = page.extract_table()
         if table:
             data.extend(table)
   
    
   df = pd.DataFrame(data)
   df.to_csv(f'{pdf_files[n]}.csv', index = False)
   n = n+1
```
