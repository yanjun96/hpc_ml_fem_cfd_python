```python
import numpy as np
import matplotlib.pyplot as plt

# Define grid
N = 10
x = np.linspace(0, 1, N+1)
vh_values = np.array([0, 0.6, 0.2, 0.5, 0.3, np.nan, np.nan, np.nan, 0.4, 0.6, 0.2])  # Manually crafted values

# Plot figure
fig, ax = plt.subplots(figsize=(10, 4))

# Plot vertical and horizontal axis
ax.axhline(0, color='black')
ax.axvline(0, color='black')

# Mark and label grid points
for i, xi in enumerate(x):
    ax.plot([xi], [0], 'ko', markersize=4)
    if i == 0:
        ax.text(xi - 0.015, -0.05, r'$0 = x_0$', color='teal', ha='center')
    elif i == N:
        ax.text(xi, -0.05, r'$x_N = 1$', color='teal', ha='center')
    else:
        ax.text(xi, -0.05, rf'$x_{i}$', ha='center')

# Plot piecewise linear function (with dotted extension)
valid = ~np.isnan(vh_values)
ax.plot(x[valid], vh_values[valid], 'k.-', linewidth=2, markersize=6)

# Extension with dotted line
ax.plot(x[4:6], vh_values[4:6], 'k:', linewidth=2)
ax.plot(x[8:], vh_values[8:], 'k:', linewidth=2)

# Label function value at x3
ax.text(x[3], vh_values[3] + 0.05, r'$v_h(x_3)$', color='teal')

# Add top text
ax.text(0.05, 1.05, r'$v_h$ is determined by $\{v_h(x_i)\}_{i=0}^N$ for points $x_i = ih$.', 
        transform=ax.transAxes, fontsize=13, 
        color='black')

# Highlight x_i = ih in red
ax.text(0.75, 1.05, r'$x_i = ih$', transform=ax.transAxes, fontsize=13, color='red')

# Formatting
ax.set_xlim(-0.05, 1.05)
ax.set_ylim(-0.1, 0.8)
ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# X-axis positions for nodes
x_vals = np.arange(-2, 2)
x_points = x_vals  # xi-3 to xi+1

# Create hat functions
x_hat1 = [x_vals[0], x_vals[1], x_vals[2]]  # phi_{i-1}
y_hat1 = [0, 1, 0]

x_hat2 = [x_vals[1], x_vals[2], x_vals[3]]  # phi_i
y_hat2 = [0, 1, 0]

# Plot
fig, ax = plt.subplots(figsize=(6, 4))

# Axes
ax.axhline(0, color='black')
ax.axvline(-2, color='black')

# Horizontal line at y=1
ax.hlines(1, x_vals[0] - 0.5, x_vals[-1] + 0.5, linestyles='dotted', colors='gray')

# Plot hat functions
ax.plot(x_hat1, y_hat1, 'k:', linewidth=2)
ax.plot(x_hat2, y_hat2, 'k-', linewidth=2)

# X-axis ticks and labels
for i, xi in enumerate(x_points):
    ax.plot([xi], [0], 'k|', markersize=12, markeredgewidth=2)
    label = rf'$x_{{i{f"+{xi}" if xi > 0 else f"{xi}" if xi < 0 else ""}}}$'
    ax.text(xi, -0.1, label, color='slateblue', ha='center', fontsize=20)

# Labels for phi
ax.text(-1.1, 1.1, r'$\phi_{i-1}$', color='teal', fontsize=20)
ax.text(-0.1, 1.1, r'$\phi_i$', color='teal', fontsize=20)

# Label y=1
ax.text(x_vals[0] - 0.4, 1, r'$1$', color='teal', va='center', fontsize=16)

# Formatting
ax.set_xlim(x_vals[0] - 0.5, x_vals[-1] + 1)
ax.set_ylim(-0.2, 1.5)
ax.axis('off')

plt.tight_layout()
plt.savefig('hat.pdf')
plt.show()
```
