```python
def minimum(*args):
    """Tanks any numner of arguments"""
    m = args[0]
    for x in args[1:]:
        if x < m:
            m = x
    return m        
```

```python
c = minimum(10,2,3,500,90,6)
c
```

```python
def maximum(*args):
    """got the maximum value"""
    m = args[0]
    for i in args[1:]:
        if i > m:
            m = i
    print('The maximum value is ',m,'.')        
    return m       
```

```python
maximum(10,20,50,800)
```

```python
def fib(n):
    if n==0 or n==1:
        return n
    else:
        return fib(n-1)+fib(n-2)
```

```python
# Decorator function
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

# Using the decorator
@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
def plus1(f):
    def wrapper(*args, **kwargs):
        return f(*args, **kwargs) +1
    return wrapper
```

```python
@plus1
def power(base, x):
    return base**x
    
power(2,3)
```

```python
def plus20(f):
    def wrapper(*args, **kwargs):
        return f(*args, **kwargs) + 20
    return wrapper

    
```

```python
@plus20
def power20(base, x):
    return base ** x

power20(2,3)    
```

```python
class particle(object):
    """this is a try"""

    roar = 'I am Groet'

    def __init__(self, charge, mass, position):
        self.c = charge
        self.m = mass
        self.r = position


    
    def hear(self):
        myroar = self.roar +(
            '\nMy charge is '+str(self.c) +
            '\nMe mass is '+ str(self.m))
        print(myroar)
```
