```python
def vmax(f,n):
    max_y = f(0)
    h = 1.0 /n
    for i in range(1, n+1):
        y = f(i*h)
        if y > max_y: max_y = y
    return max_y
```

```python
def g(x): return x**2 
print(vmax(g, 10))
```

```python
def ddx(f):
    h = 1.e-6
    def f_prime(x): return  (f(x+h)-f(x)) / h 
    return f_prime
```

```python
def qsort(lst):
    if lst == []: return []
    p = lst[0]
    sml = qsort( [x for x in lst[1:] if x < p] )  # lst[1:] is list items except lst[0]
    grt = qsort( [x for x in lst[1:] if x>= p] )
    return sml + [p] + grt
```

```python
from numpy import sqrt, array
def solveq(a,b,c):
    d = b**2 - 4*a*c
    if d < 0: return "No real-valued solutions"
    w = sqrt(d)
    return 1.0/(2*a) * (-b + array( [w, -w] ))   # two values, so [w, -w]
```

```python
class Parabola(Polynomial):
    def __init__(self, coefficients):
        if len(coefficients) != 3:
            print ('no parabola, the list in Parabola should contain only 3 coefficiens')
            return
        else: super().__init__(coefficients) 
            # super(). calling the parents class inside a child, or subclass

    def roots(self):
        print('To be implemented')
        return

    def __add__(self, q):
        lst = [0, 0, 0]
        for i in range(3): lst[i] += self.coeff[i] + q.coeff[i]
        return Parabola(lst)

    from numpy import sqrt, array
    def solver(self):
        a = self.coeff[0]  # should not self.coefficients, since coefficients are for new list
        b = self.coeff[1]
        c = self.coeff[2]
        
        d = b**2 - 4*a*c
        if d < 0: return "No real-valued solutions"
        w = sqrt(d)
        ans = 1.0/(2*a) * (-b + array( [w, -w] )) 
        print(f'The first solution is {ans[0]}, \nThe Second solution is {ans[1]}')
        return ans 
```
