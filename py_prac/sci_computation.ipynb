```python
def vmax(f,n):
    max_y = f(0)
    h = 1.0 /n
    for i in range(1, n+1):
        y = f(i*h)
        if y > max_y: max_y = y
    return max_y
```

```python
def g(x): return x**2 
print(vmax(g, 10))
```

```python
def ddx(f):
    h = 1.e-6
    def f_prime(x): return  (f(x+h)-f(x)) / h 
    return f_prime
```

```python
print(ddx(g)(2))
dgdx = ddx(g)
print(dgdx(2))
```

```python
def ack(x,y):
    if x == 0: return y + 1
    if y == 0: return ack(x-1, 1)
    return ack(x-1, ack(x, y-1))
```

```python
def qsort(lst):
    if lst == []: return []
    p = lst[0]
    sml = qsort( [x for x in lst[1:] if x < p] )  # lst[1:] is list items except lst[0]
    grt = qsort( [x for x in lst[1:] if x>= p] )
    return sml + [p] + grt
```

```python
class Polynomial:  
    # in the first line, it should not have (), 
    # only if it inheritance other class
    
    def __init__(self, coeff):
        self.coeff = coeff

    def __call__(self, x):
        s = 0;
        for i in range( len(self.coeff) ):  
            # i start from 0, 
            # since len(), so 'self.coeff' have to be a list
            s += self.coeff[i] * x ** i
        return s

    def __add__(self, q):
        lst = []
        if len(self.coeff) > len(q.coeff):
            lst += self.coeff  
            ######################
            # if use lst = self.coeff, it is only another name of self.coeff
            # lst += self.coeff is: lst = self.coeff.copy()
            
            for i in range(len(q.coeff)):
                lst[i] += q.coeff[i]  # not change item of q
        else:
            lst += q.coeff
            for i in range(len(self.coeff)):
                lst[i] += self.coeff[i]
        return Polynomial(lst)


    def __mul__(self, q):
        d1, d2 = len(self.coeff), len(q.coeff)
        lst = [0 for i in range(d1 + d2 - 1)]
        for i in range(d1):
            for j in range(d2):
                lst[i+j] += self.coeff[i] * q.coeff[j]
        return Polynomial(lst)

    def __eq__(self, q):
        if len(self.coeff) != len(q.coeff):
            return False
        
        for i in range( len(self.coeff)):
            if abs(self.coeff[i] - q.coeff[i]) >= 1.e-14: 
                # abs is necessary since it may be a negative value
                return False
        return True                
        
```
