```python
index = 1
while index < 4:
    print(f"{index} times 5 is {5*index}")
    index += 1
    
```

Floating precision

```python
epi = 0.1
i = 0
while 1 + epi > 1:  # epi is the machine epsilon scale
    epi /= 2
    i   += 1
print(epi)
print(i)
```

```python
epi = 0.1
i = 0
while  epi > 0:   # epi is the smallest representable float
    epi /= 2
    i   += 1
print(epi)
print(i)
```
