```python
import numpy as np
from scipy.optimize import minimize
```

```python
def objective(x):  #function to minimize
    return (x[0] - 3)**2 + 2

x0 = [0]  # starting guess

result = minimize(objective, x0)

print('optimal solution:', result.x)  # best x
print('objective value:', result.fun) # function value at that x
```

```python
def objective(x):
    return x[0]**2

def constraint(x):
    return x[0]-2

x0 = [0.5]

con = {'type': 'ineq', 'fun': constraint}
# ineq: function must be >= 0
# eq: function must be = 0

result = minimize(objective, x0, constraints=con)

print('optimal solution:', result.x)  # best x
print('objective value:', result.fun)
```
