```python
def minimum(*args):
    """Tanks any numner of arguments"""
    m = args[0]
    for x in args[1:]:
        if x < m:
            m = x
    return m        
```

```python
c = minimum(10,2,3,500,90,6)
c
```

```python
def maximum(*args):
    """got the maximum value"""
    m = args[0]
    for i in args[1:]:
        if i > m:
            m = i
    print('The maximum value is ',m,'.')        
    return m       
```

```python
maximum(10,20,50,800)
```

```python
# Decorator function
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

# Using the decorator
@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
class particle(object):
    """this is a try"""

    roar = 'I am Groet'

    def __init__(self, charge, mass, position,flavor):
        self.c = charge
        self.m = mass
        self.r = position
        self.flavor = flavor


    
    def hear(self):
        myroar = self.roar +(
            '\nMy charge is '+str(self.c) +
            '\nMe mass is '+ str(self.m))
        print(myroar)

    def flip(self):
        if self.flavor == "up":
           self.flavor = "down"
        elif self.flavor == 'down':
           self.flavor = 'up'
        return self.flavor

    # Without this @staticmethod, use c1.possible can't return.
    @staticmethod 
    def possible():
        return ['up','down','top']
```

```python
c1 = particle(1,2,'I am position, 3','up')
```

```python
c1.hear()
```

```python
c1.flip()
```

```python
print(c1.flip())
```

```python
c1.possible()
```

```python
from math import sqrt
def add_is_p(cls):
    def distance(self):
        d2 = 0
        for axis in ['x','y']:
            d2 += (self.r[axis]) **2
        d = sqrt(d2)
        return d
    cls.distance = distance
    return cls
```

```python
@add_is_p
class particle(object):
    """this is a try"""

    roar = 'I am Groet'

    def __init__(self):
        self.c = 0
        self.m = 0
        self.r = {'x':1, 'y':1}
```

```python
c = particle()
```
