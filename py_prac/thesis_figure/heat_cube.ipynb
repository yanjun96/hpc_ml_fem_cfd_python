```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Define cube vertices
c = 1
cube_vertices = np.array([[0, 0, 0],
                          [c, 0, 0],
                          [c, c, 0],
                          [0, c, 0],
                          [0, 0, c],
                          [c, 0, c],
                          [c, c, c],
                          [0, c, c]])

# Define cube faces
faces = [[cube_vertices[j] for j in [0, 1, 2, 3]],
         [cube_vertices[j] for j in [4, 5, 6, 7]],
         [cube_vertices[j] for j in [0, 1, 5, 4]],
         [cube_vertices[j] for j in [2, 3, 7, 6]],
         [cube_vertices[j] for j in [1, 2, 6, 5]],
         [cube_vertices[j] for j in [4, 7, 3, 0]]]

# Draw cube with transparent faces
cube = Poly3DCollection(faces, facecolors='lightgray', linewidths=1, edgecolors='black', alpha=0.3)
ax.add_collection3d(cube)

# Define center of cube
center = np.array([0.5, 0.5, 0.5])

b=2.5
# Define vectors (Q_x, Q_y, Q_z and deltas)
vectors = {
    r'$Q_x$':      (-b, 0, 0),
    r'$Q_x+\Delta x$': (b, 0, 0),
    r'$Q_y$':      (0, -b, 0),
    r'$Q_y+\Delta y$': (0, b, 0),
    r'$Q_z$':      (0, 0, -b),
    r'$Q_z+\Delta z$': (0, 0, b),
}

# Draw arrows and labels
for label, direction in vectors.items():
    dx, dy, dz = direction
    ax.quiver(*center, dx, dy, dz, length=1.2, color='black', normalize=True)
    ax.text(*(center + 0.55 * np.array([dx, dy, dz])), label, fontsize=20)

# Axes labels and vectors
ax.quiver(0, 0, 0, 0.4, 0, 0, color='black')
ax.text(0.45, 0, 0, 'x', fontsize=12)

ax.quiver(0, 0, 0, 0, 0.4, 0, color='black')
ax.text(0, 0.45, 0, 'y', fontsize=12)

ax.quiver(0, 0, 0, 0, 0, 0.4, color='black')
ax.text(0, 0, 0.45, 'z', fontsize=12)

# Set limits and aspect ratio
ax.set_xlim([-0.5, 1.5])
ax.set_ylim([-0.5, 1.5])
ax.set_zlim([-0.5, 1.5])
ax.set_box_aspect([1, 1, 1])
ax.set_axis_off()

plt.tight_layout()
plt.show()
```
