```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("1.csv")
df.columns = df.iloc[0] 
df = df[1:].reset_index(drop=True)
df = df.iloc[:,:3]
df.columns = [ "t1", "t2", "t3"]
df = df.astype({'t1':float, 't2':float, 't3':float})
df['ave'] = (df['t1'] + df['t2'] + df['t3']) / 3
print(df)
```

```python
plt.plot(df['t1'])
plt.plot(df['t2'])
plt.plot(df['t3'])
```

```python
plt.plot(df['ave'])
```
