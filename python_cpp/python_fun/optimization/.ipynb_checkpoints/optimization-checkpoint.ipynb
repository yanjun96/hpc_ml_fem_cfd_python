```python
import numpy as np
from scipy.optimize import minimize
```

```python
def objective(x):  #function to minimize
    return (x[0] - 3)**2 + 2

x0 = [0]  # starting guess

result = minimize(objective, x0)

print('optimal solution:', result.x)  # best x
print('objective value:', result.fun) # function value at that x
```

```python
def objective(x):
    return x[0]**2

def constraint(x):
    return x[0]-2

x0 = [0.5]

con = {'type': 'eq', 'fun': constraint}
# ineq: function must be >= 0
# eq: function must be = 0

result = minimize(objective, x0, constraints=con)

print('optimal solution:', result.x)  # best x
print('objective value:', result.fun)
```

```python
def f4(vars):
    x, y = vars
    return (x-1)**2 + (y-2)**2

cons = {'type':'eq', 'fun': lambda v:v[0]+v[1]-3}

res = minimize(f4, x0=[0,0], constraints=cons)
print(res.x, res.fun)
```
