```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data roughly matching the trend in the image
np.random.seed(0)
Ti = np.random.uniform(-17.5, -2.5, 50)  # Temperature range
noise = np.random.normal(0, 0.2, Ti.shape)
sigma_s = -0.076 * Ti + 0.179 + noise  # Linear trend with noise

# Linear regression
slope, intercept = np.polyfit(Ti, sigma_s, 1)
r_squared = np.corrcoef(Ti, sigma_s)[0, 1] ** 2

# Plotting
plt.figure(figsize=(6, 4))
plt.scatter(Ti, sigma_s, color='blue', s=20)

# Regression line
Ti_line = np.linspace(min(Ti), max(Ti), 100)
sigma_line = slope * Ti_line + intercept
plt.plot(Ti_line, sigma_line, 'k-', linewidth=1)

# Labels
plt.xlabel(r'Ice temperature $T_{ice}$ / $^\circ$C', fontsize=16)
plt.ylabel(r'Ice shear strength (MPa)', fontsize=16)
plt.xlim(-18, -2)
plt.ylim(0, 2)


plt.tight_layout()
plt.savefig('shear_ice.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data roughly mimicking the plot
T_ice = np.array([-7.8, -7.2, -6.8, -6.3, -6, -5.5, -5, -4.5, -4, -3.5, -3, -2.5, -2, -1.5, -1, -0.5, 0])

P_ice = -0.4524 * T_ice + 0.279 + np.random.normal(0, 0.1, len(T_ice))  # Add slight noise

# Linear regression
slope, intercept = np.polyfit(T_ice, P_ice, 1)
r_squared = np.corrcoef(T_ice, P_ice)[0, 1] ** 2

# Plot
plt.figure(figsize=(6, 4))
plt.scatter(T_ice, P_ice, color='black', label='Compressive strength')
plt.plot(T_ice, slope * T_ice + intercept, 'r-', linewidth=2, label='Fitting curve')

# Labels and annotation
plt.xlabel(r'Ice temperature $T_{ice}$ / $^\circ$C', fontsize=16)
plt.ylabel(r'Compressive strength $\sigma$ / MPa', fontsize=16)
#plt.text(-6.5, 2.5, fr'$P_{{ice}} = {slope:.4f}T_{{ice}} + {intercept:.3f}$' + '\n' + fr'$R^2 = {r_squared:.2f}$', fontsize=12)
#plt.legend()
plt.grid(False)
plt.tight_layout()
plt.savefig('compressive_ice.pdf')
```
