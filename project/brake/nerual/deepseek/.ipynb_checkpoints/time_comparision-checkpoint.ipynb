```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4]
y = [108, 528, 0.5, 9.65]
#y = np.log(y)
plt.bar(x, y, width=0.4)

label = ['104 min','528 min','29 s','9.6 min']
# Add text on top of each bar
for i in range(len(x)):
    plt.text(x[i], y[i] + 0.2, label[i], ha='center')

font_g = 14

plt.xlabel('Methods', fontsize=font_g)
plt.ylabel('Computational time / min', fontsize=font_g)
#plt.title('Computational time compare with FEM and DNNs')
lables = ['FEM \n mesh, 5 mm \n 1 million DOFs','FEM \n mesh, 3 mm \n 4.7 million DOFs',
          'DNNs\n data shape \n 4×1000 ','DDNs \n data shape \n  4×4000 data']
plt.xticks(x, lables, fontsize=10)
#plt.yticks(y)

plt.tight_layout()
#plt.grid(True)
plt.savefig('t_compare.pdf')
```
