# 1:PDF to CSV

```python
import os
import pdfplumber
import pandas as pd

source_dir = "/Users/yanjun/Documents/courses/lectures/fdd3431/brake/data/csv/"
pdf_files = [f for f in os.listdir(source_dir) if f.endswith(".pdf") ]

data = []
n=0
for file in pdf_files:
   data = []
   with pdfplumber.open(file) as pdf:
     for page in pdf.pages:
         table = page.extract_table()
         if table:
             data.extend(table)
   n = n+1
   df = pd.DataFrame(data)
   df.to_csv(f'{n}.csv', index = False)
```

# 2: CSV to CSV_ave

```python
import os
import pandas as pd

source_dir1 = "/Users/yanjun/Documents/pdc_fem_cfd_python_cpp/lectures/fdd3431/brake/nerual/csv/"
target_dir = "/Users/yanjun/Documents/pdc_fem_cfd_python_cpp/lectures/fdd3431/brake/nerual/csv_ave/"

csv_files = [f for f in os.listdir(source_dir1) if f.endswith(".csv")]

data = []
n = 0

for file in csv_files:
    df= pd.read_csv(file)
    df.columns = df.iloc[0]
    df = df[1:].reset_index(drop = True)
    df = df.iloc[:,:3]
    df.columns = [ "t1", "t2", "t3"]
    df = df.astype({'t1':float, 't2':float, 't3':float})
    df['ave'] = (df['t1'] + df['t2'] + df['t3']) / 3
    name = os.path.splitext(os.path.basename(file))[0] + "_ave.csv"
    new_f_name = os.path.join(target_dir,  name )
    
    df.to_csv(new_f_name, index = False)
    
```
