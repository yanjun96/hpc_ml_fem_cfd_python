```python
class student:

    def __init__(self, age, score):
        self.age = age
        self.score = score
        self.money = age*score
        self.experience = age**2

    def study(self, year):
        self.future = self.money * year

    def add_years(self, new_year):
        self.age += new_year
        self.money = self.age * self.score


c = student(10,20)
print("1: age now is", c.age)
print("1: salary now is", c.money, "sek /hour")

c.add_years(2)
print("2: age after add two years is", c.age)
print("2: salary after two years is", c.money, "sek /hour")
```

```python
d = student(50, 100)

print("1: age now is", d.age)
print("1: salary now is", d.money, "sek /hour")

d.add_years(2)
print("2: age after add two years is", d.age)
print("2: salary after two years is", d.money, "sek /hour")
```
